## Optical Character Recognition (OCR)

Character detector which extracts printed or handwritten text from an image or video.
We will use pytesseract to perform this task

Importing libraries

````python
import cv2
import pytesseract
from pytesseract import Output
import matplotlib.pyplot as plt
````

We will use following stock legal document (Link: https://eforms.com/purchase-agreements/stock/) for OCR

![title](./images/sample-image.png)


````python
image = cv2.imread('./images/sample-image.png')
print("The type of this input is {}".format(type(image)))
print("Shape: {}".format(image.shape))
plt.imshow(image)
````

Optional : Converting ```BGR``` image outputted by ```cv2.imread``` to grayscale. Since, our image is of document we dont need to perform this step

````python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
plt.axis('off')
````

## Performing OCR 

````python
pytesseract.pytesseract.tesseract_cmd = 'C:/Program Files/Tesseract-OCR/tesseract.exe'  # your path may be different
````

````python
print(pytesseract.image_to_string('./images/sample-image.png'))
````

## Visualizing output

Get bounding box to visualize which part of image is recognized by pytesseract

````python
h, w, d = image.shape
boxes = pytesseract.image_to_boxes(image) 
for b in boxes.splitlines():
    b = b.split(' ')
    img = cv2.rectangle(image, (int(b[1]), h - int(b[2])), (int(b[3]), h - int(b[4])), (0, 255, 0), 1)
````

````python
plt.figure(figsize=(20,16))
plt.imshow(img)
plt.axis('off')
````
